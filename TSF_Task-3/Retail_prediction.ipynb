```python
plt.figure(figsize=(10, 10));
sns.countplot(x=df['State'])
plt.xticks(rotation=90);
```

```python
df['State'].value_counts()
```

Maximun buyers are from California and New York

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Region',data=df,palette='rocket_r',order=df['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

Most of the people are from the west region

```python
df.hist(figsize=(10,10),bins=60)
plt.show()
```

Maximum discount of 0 to 20 percent is offered and most customers tends to buy quantity of 2 and 3


```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Segment',y='Sales',data=df,hue='Segment',order=df['Segment'].value_counts().index,palette='rocket')
plt.xlabel('Segment',fontsize=15)
plt.show()
```

```python
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

the sale segment is almost same in all of the four regions

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','green'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

Highest profit is for california and new york whereas for texas, pennsylvania, Ohio its a loss.

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],alpha=0.9,figsize=(10,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

**Final Observations**

-California and New York have the highest number of buyers hence the business there is most succesful.

-Profit and sales are maximum in consumer segment and minimum in Home Office segment.

-Profit and sales are maximum in west region and minimum in south region hence it should work on central and the south regions  for better establishments.

-Furniture shows less profit whereas technology and Office supplies have higher profit
